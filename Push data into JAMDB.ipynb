```python
import requests
import os
import json
import sys

path = '/Users/chriswisecarver/Dropbox/cos-dev/jamdb/data'

jam_base_url = 'https://staging-metadata.osf.io/v1/'
jam_jwt = 'eyJ0eXAiOiJKV1QiLCJhbGciOiJIUzI1NiJ9.eyJzdWIiOiJ0cmFja2VkLW9wZW50cmlhbHN8dXNlcnMtYWRtaW4iLCJleHAiOjE4MTU2NTI1ODl9.smVVQ972oDiHZj4eXRO8GWUtfCstz_qboEitRIEGWVg'

# jam_base_url = 'http://localhost:1212/v1/'
# jam_jwt = 'eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJleHAiOjM3NDU1NzMyNjUxLCJzdWIiOiJqYW0tT1BFTlRSSUFMUzp1c2Vycy1zY3JhcGkifQ.A37dFckSIazis6VhQ3YBQUnoRIfY8Cda4MdSkPXj36M'

headers = {'Authorization': jam_jwt, 'Content-Type': 'application/vnd.api+json; supported-ext="bulk,jsonpatch"'}

count = 0
bulk_data = dict(data=[])

def filter_for_elastic(d):
    stringify_keys = ['sub_title',]
    new_data = {}
    for k, v in d.items():
        if isinstance(v, str) and k == 'detailed_description':
            v = v[:256]
        if isinstance(v, int) and v >= 9223372036854775807:
            v = str(v)
        if isinstance(v, float) and v in [float('Infinity'), float('-Infinity'), float('nan')]:
            v = str(v)
        if k in stringify_keys and isinstance(v, dict):
            new_data[k] = json.dumps(v)
        else:
            new_data[k] = v
    return new_data

for filename in os.listdir(path):
    if count <= 200:
        with open(os.path.join(path, filename)) as fp:
            temp_file_data = fp.read(None)
            data = json.loads(temp_file_data, object_hook=filter_for_elastic)
            bulk_data['data'].append(data['data'])
            count = count + 1
    else:
        try:
            req = requests.post('{}id/collections/opentrials.data/documents'.format(jam_base_url), headers=headers, json=bulk_data)
        except requests.exceptions.ConnectionError:
            req = requests.post('{}id/collections/opentrials.data/documents'.format(jam_base_url), headers=headers, json=bulk_data)
        if req.ok:
            print('Posted {} or {} records to JAM'.format(count, len(bulk_data['data'])))
            count = 0
            bulk_data['data'] = []
        else:
            print('Error: {}'.format(req.json()))
```
